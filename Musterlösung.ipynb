````python
import pandas as pd
import requests
````

````python
# einlesen von Daten aus Bibsonomy
import requests
url = "https://www.bibsonomy.org/json/search/Bread?items=1000"
url2 = "https://www.bibsonomy.org/json/search/Cheese?items=1000"

result1 = requests.get(url) 
data1 = result1.json() 

result2 = requests.get(url2) 
data2 = result2.json() 
````

Wie sehen unsere Daten aus?

````python
data1.keys()
````

````python
# data1["items"]
````

Datensätze als DataFrame:

````python
df_bread = pd.DataFrame(data1["items"])
````

````python
df_bread
````

````python
df_cheese = pd.DataFrame(data2["items"])
````

````python
df_cheese
````

Informationen zu den Spalten der Datensätze:

````python
df_cheese.info()
````

````python
df_cheese["search_term"] = "cheese"
df_bread["search_term"] = "bread"
````

````python
df_bread.columns
````

````python
# Wir betrachten ein Element (Zeile)
df_bread.loc[10]
````

````python
# eine weitere praktische Funktion: to_dict()
df_bread.loc[10].to_dict()
````

````python
# Mit .concat(Liste von DataFrames) lassen sich DataFrames miteinander kombinieren. 
df_both = pd.concat([df_bread, df_cheese])
````

````python
df_both
````

Filtern nach Typ des Eintrages:

````python
df_both = df_both.loc[df_both["type"]=="Publication"]
df_both
````

````python
# überprüfen, für welche welche Spalten in allen Zeilen Einträge vorhanden sind
df_both.dropna(axis="columns").columns
````

````python
# Spalten auswählen
df_select = df_both[['authors', 'author', 'type', 'id', 'tags', 'label', 'description', 'date', 'url', 'pub-type', 'year', 'search_term', 'isbn']]
````

````python
# Duplikate entfernen
df_select = df_select.drop_duplicates(subset = "label", keep = "first")
````

````python
df_select
````

````python
# Umbenennen von Spalten
df_select = df_select.rename(columns={"label":"title"})
````

````python
df_select
````

### Filtern und auswählen

````python
df_select.year.unique()
````

** RegEx, Reguläre Ausdrücke für Auswahl von Text**

````python
import re
````

````python
def find_number(text):
    return re.search("\d{4}", text).group(0)
````

````python
print(find_number('{2015}'))
````

````python
df_select["year"] = df_select.apply(lambda row: find_number(row['year'])  , axis=1) 
````

````python
df_select.year.unique()
````

````python
# Plot by publications per year 

df_select.year.astype(int).value_counts().sort_index().plot(kind='bar', xlabel='year', ylabel='frequency')
````

````python
# reindex, drop deletes old index, that was saves a a new column
df_select = df_select.reset_index(drop=True)
````
